Setup Data for 0.04 Mass Fraction H20

```python
eruption_file_004 = open("/Users/wyattsieminski/Desktop/EEPS 1960Z/EEPS 1960Z Final Project/eruption_h20_004.txt", "r")
eruption_string_004 = eruption_file_004.read()

eruption_004 = eruption_string_004.split(",", -1)

time_004 = []
pressure_004 = []
volume_004 = []
density_004 = []
mass_fraction_bubbles_004 = []
mass_erupted_004 = []
over_pressure_004 = []
volume_percent_bubbles_004 = []

volume_erupted_004 = []

rows = len(eruption_004)/6

lith_pressure = 9.81*5000*2500

for i in range(0, int(rows/1000)):
    time_004.append(float(eruption_004[6000*i])*3.17098e-8)
    pressure_004.append(float(eruption_004[6000*i + 1]))
    over_pressure_004.append(pressure_004[i]-lith_pressure)
    volume_004.append(float(eruption_004[6000*i + 2]))
    density_004.append(float(eruption_004[6000*i + 3]))
    mass_fraction_bubbles_004.append(float(eruption_004[6000*i + 4]))
    
    volume_percent_bubbles_004.append((((mass_fraction_bubbles_004[i]*volume_004[i]*density_004[i])/(pressure_004[i]/(462*1073.15)))/volume_004[i])*100)
    
    mass_erupted_004.append(float(eruption_004[6000*i + 5]))
    volume_erupted_004.append(mass_erupted_004[i]/density_004[i])

total_volume_erupted_004 = [volume_erupted_004[0]/1e9]

for i in range(1, len(volume_erupted_004)):
    total_volume_erupted_004.append(total_volume_erupted_004[i-1] + ((time_004[i]-time_004[i-1])*(volume_erupted_004[i])*3.154e7)/1e9)

```

Data setup for analytical solution

```python
density = 2200

shape_factor = 0.1
radius = 15.0
conduit_cross_sectional_area = np.pi*(radius**2)
height = 5000.0
dynamic_viscosity = 10**7


l = (density*shape_factor*(conduit_cross_sectional_area**2))/(height*dynamic_viscosity)
beta = (1/((1/10**10) + (1/density)*0.000015))

V = 10e9

time_analytical = np.logspace(1, 9, num=60, base=10)
overpressure_analytical = []

p0 = 1e7

for t in time_analytical:
    overpressure_analytical.append(p0*(math.e**((-1)*((l*beta*t)/(V*density)))))

for i in range(0, 60):
    time_analytical[i] = time_analytical[i]*3.17098e-8
```

```python
plt.plot(time_analytical, overpressure_analytical, 'o')
plt.xscale('log')
plt.xlabel("Time (yr)")
plt.ylabel("Overpressure (Pa)")
plt.title("Overpressure vs Time")
plt.legend(["Analytical Result"])
plt.xlim(1.0e-2, 10)
plt.grid()
plt.show()
```

```python
plt.plot(time_004, over_pressure_004, label='004 data')
plt.plot(time_analytical, overpressure_analytical, 'o', color='green')
plt.xscale('log')
plt.xlabel("Time (yr)")
plt.ylabel("Overpressure (Pa)")
plt.title("Overpressure vs Time")
plt.legend(["Numerical Model", "Analytical Solution Presented in Huppert and Woods(2002)"], loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=2)
plt.xlim(1.0e-2, 10)
plt.grid()
plt.show()
```
